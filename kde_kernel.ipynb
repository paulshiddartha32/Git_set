```python
import pandas as pd
import glob
colnames = ['Type','X','Y','Z']
df = pd.DataFrame(columns=colnames)
files = glob.glob('./Simulation_Data/*.xyz')
for filename in files:
   temp = pd.read_csv(filename, skiprows=3, names=colnames, sep=" ")
   df = df.append(temp, ignore_index=True)
df
colnames = ['Type','X','Y','Z']
df2 = df[df.Type == 2]
df2
   
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors
 
# set seaborn style
sns.set_style("white")



# Custom the color, add shade and bandwidth
sns.kdeplot(x=df2.X, y=df2.Y, cmap="viridis", shade=True, bw_adjust=.5,cbar=True, Z = [2e-5, 4e-5])
plt.xlim(2, 130)
plt.ylim(2, 225)
plt.axis('equal')
plt.show()

# sns.kdeplot(x=df2.X, y=df2.Y, cmap="viridis", shade=True, bw_adjust=.5,cbar=True)
# plt.xlim(2, 130)
# plt.ylim(2, 225)
# plt.axis('equal')
# plt.show()
```

```python
from scipy.stats import gaussian_kde
import numpy as np
x = df2.X.to_numpy().astype(float)
y = df2.Y.to_numpy().astype(float)
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
```

```python

import seaborn as sns
import numpy as np
from matplotlib.cm import ScalarMappable
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors
fig, ax = plt.subplots()
#ax.scatter(x, y, c=z, s=10, cmap = 'viridis')
x_sorted = np.sort(x)
y_sorted = np.sort(y)
X, Y = np.meshgrid(x, y)
vmin = 2e-5
vmax = 4e-5
```

```python
fig, ax = plt.subplots()
qcs = ax.contourf(X, Y, z, cmap = "viridis")
fig.colorbar(
   ScalarMappable(norm=qcs.norm, cmap=qcs.cmap),
   ticks=range(vmin, vmax, 1e-5)
)
plt.axis('equal')
plt.show()
```
